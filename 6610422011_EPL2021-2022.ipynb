## 1. What is an average of Goal score by home team and away team per match.

### Home Team

```python
mean_goal_score_home= df.groupby("HomeTeam").agg(
    hometeam_goal_score=pd.NamedAgg(column="FTHG", aggfunc="mean"),
    )
mean_goal_score_home.sort_values(by=['hometeam_goal_score'], ascending= False)
```

### Away Team

```python
mean_goal_score_away= df.groupby("AwayTeam").agg(
    awayteam_goal_score=pd.NamedAgg(column="FTAG", aggfunc="mean"),
    )
mean_goal_score_away.sort_values(by=['awayteam_goal_score'], ascending= False)
```

## 2. The percentage of home wins and away wins.

```python
hometeamwin = df[df['FTR']=='H'].groupby("HomeTeam").size()
teamsize = df.groupby("HomeTeam").size()
winrate = hometeamwin / teamsize
winrate = pd.DataFrame(winrate,columns= ['win rate(%)'])
winrate.sort_values(by='win rate(%)',ascending= False)*100

```

### Away Team

```python
awayteamwin = df[df['FTR']=='A'].groupby("AwayTeam").size()
awayteamsize = df.groupby("AwayTeam").size()
winrate = awayteamwin / awayteamsize
winrate = pd.DataFrame(winrate,columns= ['win rate(%)'])
winrate.sort_values(by='win rate(%)',ascending= False)*100
```

## 3.What is an average Number of shot, Shot on target by Home team and Away team 

### Home team.

```python
mean_shot_home= df.groupby("HomeTeam").agg(
    average_hometeam_shot=pd.NamedAgg(column="HS", aggfunc="mean"),
    average_hometeam_shot_on_target=pd.NamedAgg(column="HST", aggfunc="mean"),
    )
mean_shot_home.sort_values(by=['average_hometeam_shot','average_hometeam_shot_on_target'], ascending= False)
```

### Away team

```python
mean_shot_away= df.groupby("AwayTeam").agg(
    average_awayteam_shot=pd.NamedAgg(column="AS", aggfunc="mean"),
    average_awayteam_shot_on_target=pd.NamedAgg(column="AST", aggfunc="mean"),
    )
mean_shot_away.sort_values(by=['average_awayteam_shot','average_awayteam_shot_on_target'], ascending= False)
```

## 4. Top 10 team has the most shot accuracy(%). 

### Home Team

```python
sum_shot_on_target= df.groupby("HomeTeam").agg(sum_hometeam_shot_on_target=pd.NamedAgg(column="HST", aggfunc="sum"))
hometeam_goal = df.groupby("HomeTeam").agg(sum_hometeam_goal = pd.NamedAgg(column='FTHG', aggfunc='sum'))
shot_accuracy = ((hometeam_goal['sum_hometeam_goal'])/(sum_shot_on_target['sum_hometeam_shot_on_target']))*100
shot_accuracy.sort_values(ascending= False).head(10)
```

### Away Team

```python
sum_shot_on_target= df.groupby("AwayTeam").agg(sum_away_shot_on_target=pd.NamedAgg(column="AST", aggfunc="sum"))
awayteam_goal = df.groupby("AwayTeam").agg(sum_away_goal = pd.NamedAgg(column='FTAG', aggfunc='sum'))
shot_accuracy_away = ((awayteam_goal['sum_away_goal'])/(sum_shot_on_target['sum_away_shot_on_target']))*100
shot_accuracy_away.sort_values(ascending= False).head(10)
```
